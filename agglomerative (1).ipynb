```python
#importing libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#importing dataset (replace this path with your dataset's path)
df = pd.read_csv(r"C:\Users\admin\Downloads\breast-cancer-wisconsin.csv")
```

```python
df
```

```python
#add your dataset column names only if your dataset is not provided with column names
df.columns=["sample_code","clump","cellsize","cellshape","adhesion","epithelial","bare_nuclei","chromatin","nucleoloi","mitoses","class"]
```

```python
df
```

```python
#we are scaling the dataset
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
df
```

```python
#importing library
from sklearn.cluster import AgglomerativeClustering

#you can choose the number of clusters
aggC = AgglomerativeClustering(n_clusters=3, linkage='ward')
kY = aggC.fit_predict(df)
```

```python
aggC.labels_
```

```python
#plotting dendogram (hierarchial diagram)
import scipy.cluster.hierarchy as sch
Z = sch.linkage(df, method = 'median')
den = sch.dendrogram(Z)
plt.title('Dendrogram for the clustering of the dataset')
plt.show()
```
