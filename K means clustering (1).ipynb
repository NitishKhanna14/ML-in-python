```python
#correlation matrix to find the relationship between features
df.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
```

```python
#to convert the string to integers (use only on categorical data)
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
```

```python
df
```

```python
from sklearn.cluster import KMeans

#applying kmeans algorithm and you can choose the number of clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
```

```python
labels = kmeans.predict(df)
centroids = kmeans.cluster_centers_
```

```python
centroids
```

```python
kmeans.labels_
```

```python
#lower the inertia, better is the algorithm

kmeans.inertia_
```

```python
df2=df.iloc[:,0]
df3=df.iloc[:,6]
```

```python
plt.scatter(df2,df3,c=labels) 
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#take any 2 columns
scatter = ax.scatter(df['lug_boot'],df['class'],
                     c=kmeans.labels_,s=50)
ax.set_title('Kmeans Clustering')

plt.colorbar(scatter)
```
