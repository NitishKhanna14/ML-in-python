```python
#to convert the string to integers (use only on categorical data)
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
```

```python
#importing libraries
from sklearn.cluster import MeanShift

clustering = MeanShift(bandwidth=2, min_bin_freq=1).fit(df)
```

```python
labels = clustering.predict(df)

labels
```
