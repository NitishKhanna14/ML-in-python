```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#importing dataset (replace thispath with your dataset's path)
df = pd.read_csv(r"C:\Users\admin\Downloads\car.csv")
```

```python
#add your dataset column names only if your dataset is not provided with column names

df.columns=["buying","maint","doors","persons","lug_boot","safety","class"]
```

```python
#describing dataset (not necessary)
df.describe()
```

```python
#correlation matrix to find the relationship between attributes
df.apply(lambda x : pd.factorize(x)[0]).corr(method='pearson', min_periods=1)
```

```python
#to convert the string to integers (use only on categorical data)
from sklearn import preprocessing 
le = preprocessing.LabelEncoder()
df = df.apply(le.fit_transform)
```

```python
df
```

```python
#importing algorithm
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

#let's consider 'a' to be no. of columns, then it's --> train_test_split(df.iloc[:,0:a-1],df.iloc[:,a-1:],random_state=1)
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:6],df.iloc[:,6:],random_state=1)

clf=DecisionTreeClassifier()
clf.fit(x_train,y_train)
```

```python
#predicting output

y_test_pred=clf.predict(x_test)
```

```python
#finding accuracy

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_test_pred)
```

```python
#finding confusion matrix

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_test_pred)
```

```python
#finding f1_score

from sklearn.metrics import f1_score

f1_score(y_test, y_test_pred, average='macro')
```

```python
#finding precision, recall and f_score altogether

from sklearn.metrics import precision_recall_fscore_support

precision_recall_fscore_support(y_test, y_test_pred, average='macro')
```
